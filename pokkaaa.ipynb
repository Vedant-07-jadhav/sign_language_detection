```python
df = pd.read_csv('Data/data.csv')
dict1 = dict(zip(df["video_id"], df["bbox"]))
for video_id, bbox in dict1.items():
    video_id = int(video_id)
    bbox = bbox.strip('[]')  # Remove '[' and ']'
    bbox = bbox.split(',')   # Split by comma

    # Convert each element to int and strip spaces
    bbox = [int(x.strip()) for x in bbox]
    print(video_id)
    print(bbox)
    break
```
